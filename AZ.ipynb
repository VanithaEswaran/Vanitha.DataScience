```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
label_encoder=LabelEncoder()
```

```python
df_az.columns
```

```python
df_az['LocationID']=label_encoder.fit_transform(df_az['LocationID'])
df_az['Class']=label_encoder.fit_transform(df_az['Class'])
df_az['Race_Ethnicity']=label_encoder.fit_transform(df_az['Race_Ethnicity'])
df_az['Gender']=label_encoder.fit_transform(df_az['Gender'])
df_az['Question']=label_encoder.fit_transform(df_az['Question'])
df_az['Topic']=label_encoder.fit_transform(df_az['Topic'])
```

```python
featrues=['LocationID','Race_Ethnicity','Gender','Question','Topic']
target='Class'
```

```python
x=df_az[featrues]
y=df_az[target]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model=RandomForestClassifier(n_estimators=100,random_state=42)
rf_model.fit(x_train,y_train)
```

```python
y_pr=rf_model.predict(x_test)
```

```python
acc=accuracy_score(y_test,y_pr)
acc
```
